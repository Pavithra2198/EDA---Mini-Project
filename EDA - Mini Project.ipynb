```python
def str2number(amount):
    if amount[-1] == 'M':
        return float(amount[1:-1])*1000000
    elif amount[-1] == 'K':
        return float(amount[1:-1])*1000
    else:
        return float(amount[1:])
    
fifa['Value'] = fifa['Value'].apply(lambda x: str2number(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: str2number(x))
fifa['Release Clause']=fifa['Release Clause'].apply(lambda x: str2number(x))
```
